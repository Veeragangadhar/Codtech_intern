```python
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# ========== CONFIG ==========
API_KEY = "391d38660aeb1deb160113cefd6edefa"
CITY = "London"
BASE_URL = "https://api.openweathermap.org/data/2.5/forecast"
```

```python
params = {
    "q": CITY,
    "appid": API_KEY,
    "units": "metric"
}
```

```python
response = requests.get(BASE_URL, params=params)

if response.status_code == 200:
    data = response.json()
else:
    print("Error fetching data:", response.status_code, response.text)
    exit()
```

```python
# Extract relevant forecast data
dates, temps, humidity, wind_speed = [], [], [], []

for forecast in data["list"]:
    dt = datetime.datetime.fromtimestamp(forecast["dt"])
    dates.append(dt)
    temps.append(forecast["main"]["temp"])
    humidity.append(forecast["main"]["humidity"])
    wind_speed.append(forecast["wind"]["speed"])
```

```python
plt.figure(figsize=(14, 8))
```

```python
# Temperature plot
plt.subplot(3, 1, 1)
sns.lineplot(x=dates, y=temps, marker="o", color="red")
plt.title(f"Weather Forecast for {CITY}")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
```

```python
# Humidity plot
plt.subplot(3, 1, 2)
sns.lineplot(x=dates, y=humidity, marker="o", color="blue")
plt.ylabel("Humidity (%)")
plt.xticks(rotation=45)
```

```python
# Wind speed plot
plt.subplot(3, 1, 3)
sns.lineplot(x=dates, y=wind_speed, marker="o", color="green")
plt.ylabel("Wind Speed (m/s)")
plt.xlabel("Date & Time")
plt.xticks(rotation=45)
```

```python
plt.tight_layout()
plt.show()
```
